Create app.py

```python
%%writefile app.py
import streamlit as st
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

st.set_page_config(page_title="Personalized News Summarization System")

st.title("📰 Personalized News Summarization System")

# Sidebar
st.sidebar.header("Settings")

rag_option = st.sidebar.selectbox(
    "Select RAG",
    ["Dense", "Hybrid", "Fusion"]
)

llm_option = st.sidebar.selectbox(
    "Select LLM",
    ["GPT", "Llama", "Qwen"]
)

query = st.text_input("Enter Query")

if st.button("Generate Summary"):

    st.subheader("Generated Summary")
    st.write("Summary will appear here")

    st.subheader("Retrieved Documents")
    st.write("Retrieved documents will appear here")

    st.subheader("Historical Metrics")
    st.write("Metrics will appear here")

    st.subheader("Comparison with Winner")
    st.write("Comparison results will appear here")

    st.subheader("Top Combination")
    st.success(f"Best Combination: {rag_option} + {llm_option}")

    st.subheader("Ranking Chart")

    sample = pd.DataFrame({
        "Combination":["Dense+GPT","Hybrid+GPT","Fusion+Llama"],
        "Score":[88,91,86]
    })

    fig, ax = plt.subplots()
    ax.bar(sample["Combination"], sample["Score"])
    st.pyplot(fig)

    st.subheader("Heatmap")

    heat = pd.DataFrame(
        [[0.88,0.91,0.86]],
        columns=["Dense","Hybrid","Fusion"]
    )

    fig2, ax2 = plt.subplots()
    sns.heatmap(heat, annot=True, ax=ax2)
    st.pyplot(fig2)
```
